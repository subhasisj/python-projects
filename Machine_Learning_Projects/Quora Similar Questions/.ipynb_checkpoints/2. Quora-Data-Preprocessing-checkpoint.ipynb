#### Remove HTML Tags

```python
# Remove HTML Tags
def remove_html(text):
    soup = BeautifulSoup(text,'lxml')
    html_free_text = soup.get_text()
    return html_free_text
```

#### Remove Punctuations

```python
# Remove Punctuations
def punctuation_remover(text):
    punctuation_free_text = "".join([char for char in text if char \
                                    not in string.punctuation])
    return punctuation_free_text
```

#### Word Stemming

```python
# Stemming
stemmer = PorterStemmer()

def word_stemmer(text):
#     text = text.split()
    stemmed_text = " ".join(stemmer.stem(word) for word in text)
    return stemmed_text
```

#### Stop Word Removal

```python
# Stop Word Removal
cached_stop_words = stopwords.words('english') # Provides 70 X Speedup
def stop_words_remover(text):
    text = text.split()
    words = [word for word in text if \
             word not in cached_stop_words]
    return words
```

#### Text Preprocessor

```python
# lets create a preprocessing function
def text_prepocressing(text,
                       remove_stop_words=True,
                       stem_words=False
                      ):
    import re
    text = text.lower() #Lower Case
    text = remove_html(text) #Remove HTML Tags
     # Clean the text
    text = re.sub(r"[^A-Za-z0-9]", " ", text)
    text = re.sub(r"what's", "", text)
    text = re.sub(r"What's", "", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"can't", "cannot ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"I'm", "I am", text)
    text = re.sub(r" m ", " am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"60k", " 60000 ", text)
    text = re.sub(r" e g ", " eg ", text)
    text = re.sub(r" b g ", " bg ", text)
    text = re.sub(r"\0s", "0", text)
    text = re.sub(r" 9 11 ", "911", text)
    text = re.sub(r"e-mail", "email", text)
    text = re.sub(r"\s{2,}", " ", text)
    text = re.sub(r"quikly", "quickly", text)
    text = re.sub(r" usa ", " America ", text)
    text = re.sub(r" USA ", " America ", text)
    text = re.sub(r" u s ", " America ", text)
    text = re.sub(r" uk ", " England ", text)
    text = re.sub(r" UK ", " England ", text)
    text = re.sub(r"india", "India", text)
    text = re.sub(r"switzerland", "Switzerland", text)
    text = re.sub(r"china", "China", text)
    text = re.sub(r"chinese", "Chinese", text) 
    text = re.sub(r"imrovement", "improvement", text)
    text = re.sub(r"intially", "initially", text)
    text = re.sub(r"quora", "Quora", text)
    text = re.sub(r"([0-9]+)000000", r"\1m", text) # 1 million
    text = re.sub(r"([0-9]+)000", r"\1k", text) # 1 Thousand
    
    # remove comma between numbers, i.e. 15,000 -> 15000
    text = re.sub('(?<=[0-9])\,(?=[0-9])', "", text)

    text = punctuation_remover(text) 
    
#     if remove_stop_words:
#         text = stop_words_remover(text)
    
    if stem_words:
        text = word_stemmer(text)

    return text
```

#### Advanced Feature Extraction (NLP & Fuzzy Features)

https://chairnerd.seatgeek.com/fuzzywuzzy-fuzzy-string-matching-in-python/

Definition:
- __Token__: You get a token by splitting sentence a space
- __Stop_Word__ : stop words as per NLTK.
- __Word__ : A token that is not a stop_word


Features:
- __cwc_min__ :  Ratio of common_word_count to min lenghth of word count of Q1 and Q2 <br>cwc_min = common_word_count / (min(len(q1_words), len(q2_words))
<br>
<br>
- __cwc_max__ :  Ratio of common_word_count to max lenghth of word count of Q1 and Q2 <br>cwc_max = common_word_count / (max(len(q1_words), len(q2_words))
<br>
<br>
- __csc_min__ :  Ratio of common_stop_count to min lenghth of stop count of Q1 and Q2 <br> csc_min = common_stop_count / (min(len(q1_stops), len(q2_stops))
<br>
<br>
- __csc_max__ :  Ratio of common_stop_count to max lenghth of stop count of Q1 and Q2<br>csc_max = common_stop_count / (max(len(q1_stops), len(q2_stops))
<br>
<br>
- __ctc_min__ :  Ratio of common_token_count to min lenghth of token count of Q1 and Q2<br>ctc_min = common_token_count / (min(len(q1_tokens), len(q2_tokens))
<br>
<br>

- __ctc_max__ :  Ratio of common_token_count to max lenghth of token count of Q1 and Q2<br>ctc_max = common_token_count / (max(len(q1_tokens), len(q2_tokens))
<br>
<br>
        
- __last_word_eq__ :  Check if First word of both questions is equal or not<br>last_word_eq = int(q1_tokens[-1] == q2_tokens[-1])
<br>
<br>

- __first_word_eq__ :  Check if First word of both questions is equal or not<br>first_word_eq = int(q1_tokens[0] == q2_tokens[0])
<br>
<br>
        
- __abs_len_diff__ :  Abs. length difference<br>abs_len_diff = abs(len(q1_tokens) - len(q2_tokens))
<br>
<br>

- __mean_len__ :  Average Token Length of both Questions<br>mean_len = (len(q1_tokens) + len(q2_tokens))/2
<br>
<br>

- __token_set_ratio__ : https://github.com/seatgeek/fuzzywuzzy#usage

<br>
<br>





- __longest_substr_ratio__ :  Ratio of length longest common substring to min lenghth of token count of Q1 and Q2<br>longest_substr_ratio = len(longest common substring) / (min(len(q1_tokens), len(q2_tokens))


```python

def get_token_features(ques_1,ques_2):
    SAFE_DIV = 0.0001 
    tokenizer = RegexpTokenizer(r'\w+')
    
    # Convert to tokens
    tokenize_strings = (lambda x : tokenizer.tokenize(x.lower()))
    ques_1_tokens = tokenize_strings(ques_1)
    ques_2_tokens = tokenize_strings(ques_2)
#     ques_2_tokens = ques_2.apply(lambda x : tokenizer.tokenize(x.lower()))
    
    # Return 0.0 for all 10 token features if tokens are empty
    token_features = [0.0] * 10
    if len(ques_1_tokens)==0 or len(ques_2_tokens)==0:
        return token_features
    
    # Get non stop words
    ques_1_words = set(stop_words_remover(ques_1))
    ques_2_words = set(stop_words_remover(ques_2))
    
    # Get stop words
    ques_1_stops = set([word for word in ques_1 if word in cached_stop_words])
    ques_2_stops = set([word for word in ques_2 if word in cached_stop_words])
    
    # Count of common words  
    common_words_count = len(ques_1_words.intersection(ques_2_words))
    
    # Count of common stop words
    common_stop_words_count = len(ques_1_stops.intersection(ques_2_stops))
    
    # Count of common tokens
    common_token_count = len(set(ques_1_tokens).intersection(set(ques_2_tokens)))

    #cwc_min
    token_features[0] = common_words_count / (min(len(ques_1_words),len(ques_2_words)) + SAFE_DIV )
    #cwc_max
    token_features[1] = common_words_count / (max(len(ques_1_words),len(ques_2_words)) + SAFE_DIV)
    
    #csc_min
    token_features[2] = common_stop_words_count / (min(len(ques_1_stops),len(ques_2_stops))+ SAFE_DIV)
    #csc_max
    token_features[3] = common_stop_words_count / (max(len(ques_1_stops),len(ques_2_stops))+ SAFE_DIV)
    
    #ctc_min
    token_features[4] = common_token_count / (min(len(ques_1_tokens),len(ques_2_tokens))+ SAFE_DIV)
    #ctc_max
    token_features[5] = common_token_count / (max(len(ques_1_tokens),len(ques_2_tokens))+ SAFE_DIV)
    
    #last_word_eq
    token_features[6] = int(ques_1_tokens[-1]==ques_2_tokens[-1])
    #first_word_eq
    token_features[7] = int(ques_1_tokens[0]==ques_2_tokens[0])

    #abs_len_diff
    token_features[8] = abs(len(ques_1_tokens) - len(ques_2_tokens))
    #mean_len
    token_features[9] = np.mean([len(ques_1_tokens) , len(ques_2_tokens)])
    
    return token_features

# Get Longest common substring ratio
def get_longest_substring_ratio(ques_1,ques_2):
    strs = list(distance.lcsubstrings(ques_1,ques_2))
    if len(strs) == 0:
        return 0
    else:
        return len(strs[0]) / (min(len(ques_1),len(ques_2)) + 1)
    
def extract_advanced_features(df):
     # preprocessing each question
    df["question1"] = df["question1"].fillna("").apply(text_prepocressing)
    df["question2"] = df["question2"].fillna("").apply(text_prepocressing)

    print("token features...")
    
    # Merging Features with dataset
    
    token_features = df.apply(lambda x: get_token_features(x["question1"], x["question2"]), axis=1)
    
    df["cwc_min"]       = list(map(lambda x: x[0], token_features))
    df["cwc_max"]       = list(map(lambda x: x[1], token_features))
    df["csc_min"]       = list(map(lambda x: x[2], token_features))
    df["csc_max"]       = list(map(lambda x: x[3], token_features))
    df["ctc_min"]       = list(map(lambda x: x[4], token_features))
    df["ctc_max"]       = list(map(lambda x: x[5], token_features))
    df["last_word_eq"]  = list(map(lambda x: x[6], token_features))
    df["first_word_eq"] = list(map(lambda x: x[7], token_features))
    df["abs_len_diff"]  = list(map(lambda x: x[8], token_features))
    df["mean_len"]      = list(map(lambda x: x[9], token_features))
   
    #Computing Fuzzy Features and Merging with Dataset
    
    # do read this blog: http://chairnerd.seatgeek.com/fuzzywuzzy-fuzzy-string-matching-in-python/
    # https://stackoverflow.com/questions/31806695/when-to-use-which-fuzz-function-to-compare-2-strings
    # https://github.com/seatgeek/fuzzywuzzy
    print("fuzzy features..")

    df["token_set_ratio"]       = df.apply(lambda x: fuzz.token_set_ratio(x["question1"], x["question2"]), axis=1)
    # The token sort approach involves tokenizing the string in question, sorting the tokens alphabetically, and 
    # then joining them back into a string We then compare the transformed strings with a simple ratio().
    df["token_sort_ratio"]      = df.apply(lambda x: fuzz.token_sort_ratio(x["question1"], x["question2"]), axis=1)
    df["fuzz_ratio"]            = df.apply(lambda x: fuzz.QRatio(x["question1"], x["question2"]), axis=1)
    df["fuzz_partial_ratio"]    = df.apply(lambda x: fuzz.partial_ratio(x["question1"], x["question2"]), axis=1)
    df["longest_substr_ratio"]  = df.apply(lambda x: get_longest_substring_ratio(x["question1"], x["question2"]), axis=1)
    return df
```

#### Visualizing Higher Dimension Space with T-SNE

```python
# Only considering the engineered features
X = sample_data_for_tsne.iloc[:,6:]
X.head()
```

```python
y = sample_data_for_tsne.is_duplicate.values
y
```

##### Build T-SNE with 2 Components

```python
tsne_2_dimensional = TSNE(n_components=2,
                          init='random',
                          random_state=45,
                          n_iter=1000,
                          verbose=2,
                         ).fit_transform(X)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
                      x = tsne_2_dimensional[:,0],
                      y = tsne_2_dimensional[:,1],
                      mode = 'markers',
                      text = y,
                      marker=dict(
                            color=y,
                            colorscale='Viridis',
                            line_width=1
                            )
))
fig.update_layout(title = 't-SNE 2D Visualization')
fig.show()
```

##### T-SNE with 3 Components

```python
tsne_3_dimensional = TSNE(n_components=3,
                          init='random',
                          random_state=45,
                          n_iter=1000,
                          verbose=2,
                         ).fit_transform(X)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(
                      x = tsne_3_dimensional[:,0],
                      y = tsne_3_dimensional[:,1],
                      z = tsne_3_dimensional[:,2],
                      mode = 'markers',
                      text = y,
                      marker=dict(
                            color=y,
                            sizemode='diameter',
                            colorscale='Viridis',
                            line_width=1,
                            colorbar = dict(title = 'duplicate'),
                            line=dict(color='rgb(255, 255, 255)'),
                            opacity=0.75
                            )
))
fig.update_layout(title = 't-SNE 3D Visualization',
                  height=800, width=800)
fig.show()
```
